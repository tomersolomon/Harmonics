```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages
```

```python
pp = PdfPages('100samples.pdf')

x = np.linspace(0,2*np.pi,100)
y = np.sin(x)
y1 = np.sin(2*x)
y2 = np.sin(4*x)

fig,ax=plt.subplots()
ax.plot(x,y)
plt.title('y = sin(x)')
ax.set_xlabel('Time')
pp.savefig()

fig,ax=plt.subplots()
ax.plot(x,y1)
plt.title('y = sin(2*x)')
ax.set_xlabel('Time')
pp.savefig()

fig,ax=plt.subplots()
ax.plot(x,y2)
plt.title('y = sin(4*x)')
ax.set_xlabel('Time')
pp.savefig()

fig,ax=plt.subplots()

ax.plot(x,y+y1+y2)
plt.title('y = sin(x) + sin(2*x) + sin(4*x)')
ax.set_xlabel('Time')

pp.savefig()
pp.close()

```

```python
pp = PdfPages('25samples.pdf')

x1 = np.linspace(0,2*np.pi,25)
y3 = np.sin(x1)
y4 = np.sin(2*x1)
y5 = np.sin(4*x1)


fig,ax=plt.subplots()
ax.plot(x1,y3)
plt.title('y = sin(x)')
ax.set_xlabel('Time')
pp.savefig()


fig,ax=plt.subplots()
ax.plot(x1,y4)
plt.title('y = sin(2*x)')
ax.set_xlabel('Time')
pp.savefig()


fig,ax=plt.subplots()
ax.plot(x1,y5)
plt.title('y = sin(4*x)')
ax.set_xlabel('Time')
pp.savefig()



fig,ax=plt.subplots()

ax.plot(x1,y3+y4+y5)
plt.title('y = sin(x) + sin(2*x) + sin(4*x)')
ax.set_xlabel('Time')
pp.savefig()

pp.close()
```

```python
pp = PdfPages('sample_comparison.pdf')

fig,ax=plt.subplots()


test = [100,75,50,25]

for index,elem in enumerate(test):
    x= np.linspace(0,4*np.pi,elem)
    ax.plot(x,np.sin(x) + np.sin(2*x) + np.sin(4*x),label = str(elem) + " samples")
    ax.set_xlabel('Time')
    
handles, labels = ax.get_legend_handles_labels()
ax.legend(handles, labels, prop={'size':9})

pp.savefig()
pp.close()

```
